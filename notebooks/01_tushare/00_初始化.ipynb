# 通过Python SDK 调取数据



```python
import tushare as ts
# 这里注意， tushare版本需大于1.2.10
from configparser import ConfigParser
cfg = ConfigParser()
cfg.read('config.ini')
token = cfg.get('tushare','token')
print(token)
ts.set_token(token)

pro = ts.pro_api()

df = pro.trade_cal(exchange='', start_date='20181201', end_date='20190115', fields='exchange,cal_date,is_open,pretrade_date', is_open='0')
df
```

```python
df = pro.query('coinlist', start_date='20170101', end_date='20171231')
df.head()
```

```python
df = pro.query('coinpair', exchange='huobi', trade_date='20180802')
df.head()
```

```python
df = pro.coinexchanges(area_code='cn')
df.head()
```

```python
df = pro.coincap(trade_date='20190119')
# df.head()
# df['price']
df
```

```python
df = pro.coinbar(exchange='huobi', symbol='btcusdt', freq='1min', start_date='20180801', end_date='20180802')
```
