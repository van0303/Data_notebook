```python
# 获取 Token
from configparser import ConfigParser
cfg = ConfigParser()
cfg.read('../config.ini')
token = cfg.get('tushare', 'token')
token

# 初始化 tushare
import tushare as ts
ts.set_token(token)
pro = ts.pro_api()
```

# 日线行情

```python


df = pro.daily(ts_code="000002.SZ", start_date="20190101", end_date="20190318")
df.head(10)
```

```python
from matplotlib import pyplot as plt
plt.plot(df["trade_date"].head(10), df["open"].head(10), label="open")
plt.plot(df["trade_date"].head(10), df["close"].head(10), label="close")

plt.legend()
plt.show()
```
