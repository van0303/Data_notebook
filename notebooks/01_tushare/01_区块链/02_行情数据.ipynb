## 比特币每日市值

```python
df = pro.btc_marketcap(start_date='20190101',end_date='20190122').sort_values('date')
# df
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(df['date'], df['marketcap'])
plt.show()
```

## 比特币每日量价

```python
df = pro.btc_pricevol(start_date='20190101', end_date='20190122').sort_values('date').reset_index(drop=True)
df
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.plot(df['date'],df['price'])
plt.subplot(212)
plt.plot(df['date'],df['volume'], color='orange')
```
