```python
# 获取 Token
from configparser import ConfigParser
cfg = ConfigParser()
cfg.read('../config.ini')
token = cfg.get('tushare', 'token')
token

# 初始化 tushare
import tushare as ts
ts.set_token(token)
pro = ts.pro_api()
```

## 数字货币每日市值

```python
pro.coincap(trade_date='20190121')
```

```python
pro.coincap(trade_date='20190121', coin='BTC')
```

## 比特币每日市值

```python
df = pro.btc_marketcap(start_date='20190101',end_date='20190122').sort_values('date')
# df
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(df['date'], df['marketcap'])
plt.show()
```

## 比特币每日量价

```python
df = pro.btc_pricevol(start_date='20190101', end_date='20190122').sort_values('date').reset_index(drop=True)
df
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.plot(df['date'],df['price'])
plt.subplot(212)
plt.plot(df['date'],df['volume'], color='orange')
```
