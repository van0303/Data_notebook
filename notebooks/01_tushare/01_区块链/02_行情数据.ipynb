```python
# 获取 Token
from configparser import ConfigParser
cfg = ConfigParser()
cfg.read('../config.ini')
token = cfg.get('tushare', 'token')
token

# 初始化 tushare
import tushare as ts
ts.set_token(token)
pro = ts.pro_api()
```

## 数字货币每日市值

```python
pro.coincap(trade_date='20190121')
```

```python
pro.coincap(trade_date='20190121', coin='BTC')
```
